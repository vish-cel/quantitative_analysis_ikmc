```python
# Load the previously saved mapped data
df_mapped_birthing = pd.read_csv(r"C:\Users\CEL\Desktop\Quantitative Analysis\iKMC\Birthing_Form_Mapped_Data.csv")
```

```python
# --- Data Cleaning for Plotting ---

# 1. Clean 'brth_bsex_1' (Baby Gender)
# Remove the anomalous '1' entry
df_mapped_birthing_cleaned = df_mapped_birthing[df_mapped_birthing['brth_bsex_1'].isin(['Male', 'Female'])].copy()

# 2. Clean 'brth_otcome_1' (Outcome of newborn)
# Remove the anomalous '15.0' entry and any other numerical entries that might appear
df_mapped_birthing_cleaned = df_mapped_birthing_cleaned[
    ~df_mapped_birthing_cleaned['brth_otcome_1'].isin(['15.0']) &
    ~pd.to_numeric(df_mapped_birthing_cleaned['brth_otcome_1'], errors='coerce').notna() # Remove any other numerical strings
].copy()


# 3. Clean 'brth_weight_1' (Baby's birthweight)
# Filter out the likely outlier '9999' and values less than 500g (unrealistically low for live births, often data entry errors)
df_mapped_birthing_cleaned = df_mapped_birthing_cleaned[
    (df_mapped_birthing_cleaned['brth_weight_1'] < 9000) &
    (df_mapped_birthing_cleaned['brth_weight_1'] > 500)
].copy()

# 4. Clean 'brth_ga' (Gestational Age at birth)
# Filter out the likely outlier '99' and values below 20 weeks (extremely preterm, often data entry errors)
df_mapped_birthing_cleaned = df_mapped_birthing_cleaned[
    (df_mapped_birthing_cleaned['brth_ga'] < 60) & # Assuming anything above 60 is an error. Max 42 is typical.
    (df_mapped_birthing_cleaned['brth_ga'] >= 20)
].copy()

# 5. Clean 'brth_mage' (Maternal Age)
# Filter out the likely outlier '99' and unrealistic ages
df_mapped_birthing_cleaned = df_mapped_birthing_cleaned[
    (df_mapped_birthing_cleaned['brth_mage'] < 70) & # Assuming anything above 70 is an error
    (df_mapped_birthing_cleaned['brth_mage'] >= 10) # Assuming anything below 10 is an error
].copy()
```

```python
# --- Plotting ---

# Chart 1: Bar Chart - Distribution of Baby Gender (brth_bsex_1)
plt.figure(figsize=(8, 6))
df_mapped_birthing_cleaned['brth_bsex_1'].value_counts().plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribution of Baby Gender', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Babies', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('baby_gender_distribution.png')
```

```python
# Chart 2: Bar Chart - Mode of Birth (brth_mdbrth_1)
plt.figure(figsize=(8, 6))
df_mapped_birthing_cleaned['brth_mdbrth_1'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribution of Mode of Birth', fontsize=14)
plt.xlabel('Mode of Birth', fontsize=12)
plt.ylabel('Number of Births', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('mode_of_birth_distribution.png')
```

```python
# Chart 3: Histogram - Baby's Birthweight (brth_weight_1)
plt.figure(figsize=(8, 6))
plt.hist(df_mapped_birthing_cleaned['brth_weight_1'], bins=20, color='lightsalmon', edgecolor='black')
plt.title('Distribution of Baby\'s Birthweight (grams)', fontsize=14)
plt.xlabel('Birthweight (grams)', fontsize=12)
plt.ylabel('Number of Babies', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('baby_birthweight_histogram.png')
```

```python
# Chart 4: Bar Chart - Outcome of Newborn (brth_otcome_1)
plt.figure(figsize=(8, 6))
df_mapped_birthing_cleaned['brth_otcome_1'].value_counts().plot(kind='bar', color='plum')
plt.title('Distribution of Newborn Outcomes', fontsize=14)
plt.xlabel('Outcome', fontsize=12)
plt.ylabel('Number of Newborns', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('newborn_outcome_distribution.png')
```

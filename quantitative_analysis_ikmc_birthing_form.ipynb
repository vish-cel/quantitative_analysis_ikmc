```python
import pandas as pd
```

```python
# Load the CSV file, skipping the first 2 rows which are metadata
df_birthing = pd.read_csv(
    r"C:\Users\CEL\Desktop\Quantitative Analysis\iKMC\Birthing form_17th Sep 24 - 18th Jun 25.csv", skiprows=2)
```

```python
# Display basic information about the DataFrame
print("DataFrame Info:")
df_birthing.info()
```

```python
# Display the first few rows of the DataFrame
print("\nFirst 5 rows of the DataFrame:")
print(df_birthing.head())
```

```python
# Calculate descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:")
print(df_birthing.describe())
```

```python
# Check for missing values and their percentages
print("\nMissing Values and their Percentages:")
missing_values = df_birthing.isnull().sum()
missing_percentage = (df_birthing.isnull().sum() / len(df_birthing)) * 100
missing_df = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage': missing_percentage})
print(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False).to_markdown(numalign="left", stralign="left"))
```

```python
# Analyze unique values and their counts for selected categorical/identifier columns
print("\nUnique Values and Counts for selected columns:")
# Map the integer codes to meaningful labels or coded values, if available.
for col in ['brth_facility_sp', 'brth_bsex_1', 'brth_mdbrth_1', 'brth_plcbrth_1', 'brth_otcome_1', 'brth_rfrl_1']:
    if col in df_birthing.columns:
        print(f"\n--- {col} ---")
        print(df_birthing[col].value_counts().to_markdown(numalign="left", stralign="left"))
```

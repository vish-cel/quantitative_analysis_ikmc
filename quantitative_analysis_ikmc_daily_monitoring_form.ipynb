```python
import pandas as pd
```

```python
# Load the Daily Monitoring form data, skipping the specified number of rows
df_dmf = pd.read_csv(r"C:\Users\CEL\Desktop\Quantitative Analysis\iKMC\DMF_17th Sep '24 - 18th Jun '25.csv", skiprows=2)
```

```python
# Display the first few rows of the dataframe
print("First 5 rows of the Daily Monitoring Form Data:")
print(df_dmf.head())
```

```python
# Display information about the dataframe, including column names and data types
print("\nInformation about the Daily Monitoring Form Data:")
print(df_dmf.info())
```

```python
# Descriptive statistics for key numerical columns
print("\nDescriptive Statistics for Key Numerical Columns:")
print(df_dmf[['dmf_ivs_age', 'dmf_kmc_dur', 'dmf_kmc_dur_mthr']].describe())
```

```python
# Analyze distribution of country_id, site_id, facility_id, dmf_researcher_id
print("\nDistribution of Country ID (country_id):")
print(df_dmf['country_id'].value_counts(dropna=False))
print(df_dmf['country_id'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of Site ID (site_id):")
print(df_dmf['site_id'].value_counts(dropna=False))
print(df_dmf['site_id'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of Facility ID (facility_id):")
print(df_dmf['facility_id'].value_counts(dropna=False))
print(df_dmf['facility_id'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of Researcher ID (dmf_researcher_id):")
print(df_dmf['dmf_researcher_id'].value_counts(dropna=False))
print(df_dmf['dmf_researcher_id'].value_counts(normalize=True, dropna=False) * 100)
```

```python
# Check for duplicate entries in 'dmf_unique_id' and 'dmf_babyid'
# 'dmf_unique_id' should be unique per daily record
print("\nNumber of duplicate 'dmf_unique_id' entries:", df_dmf['dmf_unique_id'].duplicated().sum())
# 'dmf_babyid' will likely have duplicates as it's recorded daily for each baby
print("Number of duplicate 'dmf_babyid' entries:", df_dmf['dmf_babyid'].duplicated().sum())
print("Number of unique babies in DMF data (dmf_babyid):", df_dmf['dmf_babyid'].nunique())
```

```python
# Convert date columns to datetime objects and examine range
df_dmf['dmf_dof'] = pd.to_datetime(df_dmf['dmf_dof'], errors='coerce')
df_dmf['dmf_ivs_dodeath'] = pd.to_datetime(df_dmf['dmf_ivs_dodeath'], errors='coerce')
df_dmf['dmf_mvs_dodeath'] = pd.to_datetime(df_dmf['dmf_mvs_dodeath'], errors='coerce')

print("\nDate Range of 'dmf_dof':", df_dmf['dmf_dof'].min(), "to", df_dmf['dmf_dof'].max())
print("Date Range of 'dmf_ivs_dodeath':", df_dmf['dmf_ivs_dodeath'].min(), "to", df_dmf['dmf_ivs_dodeath'].max())
print("Date Range of 'dmf_mvs_dodeath':", df_dmf['dmf_mvs_dodeath'].min(), "to", df_dmf['dmf_mvs_dodeath'].max())
```

```python
# Analyze distributions of key categorical/binary variables (assuming 11/12 logic)
print("\nDistribution of 'dmf_ivs_bvs' (Baby's Vital Status):")
print(df_dmf['dmf_ivs_bvs'].value_counts(dropna=False))
print(df_dmf['dmf_ivs_bvs'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_phototherapy' (Phototherapy in last 12 hrs):")
print(df_dmf['dmf_l12_phototherapy'].value_counts(dropna=False))
print(df_dmf['dmf_l12_phototherapy'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_oxy_supp' (Oxygen Support in last 12 hrs):")
print(df_dmf['dmf_l12_oxy_supp'].value_counts(dropna=False))
print(df_dmf['dmf_l12_oxy_supp'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_cpap' (CPAP in last 12 hrs):")
print(df_dmf['dmf_l12_cpap'].value_counts(dropna=False))
print(df_dmf['dmf_l12_cpap'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_ab' (Antibiotics in last 12 hrs):")
print(df_dmf['dmf_l12_ab'].value_counts(dropna=False))
print(df_dmf['dmf_l12_ab'].value_counts(normalize=True, dropna=False) * 100)
```

```python
# Analyze missing value patterns for highly null columns
# List columns with very high percentage of missing values (e.g., >95%)
highly_missing_cols_dmf = df_dmf.columns[df_dmf.isnull().sum() / len(df_dmf) > 0.95].tolist()
print("\nColumns in DMF data with more than 95% missing values:")
for col in highly_missing_cols_dmf:
    print(f"- {col}: {df_dmf[col].isnull().sum() / len(df_dmf) * 100:.2f}% missing")

# Specific check for death-related information
print("\nMissing values for death-related columns:")
print(df_dmf[['dmf_ivs_dodeath', 'dmf_ivs_todeath', 'dmf_ivs_codeath', 'dmf_mvs_dodeath', 'dmf_mvs_todeath', 'dmf_mvs_codeath']].isnull().sum())
```

```python
# KMC breakdown: Examine who provided KMC
print("\nDistribution of 'dmf_kmc_who_sur1' (KMC by Surrogate 1):")
print(df_dmf['dmf_kmc_who_sur1'].value_counts(dropna=False))
print(df_dmf['dmf_kmc_who_sur1'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_kmc_who_sur2' (KMC by Surrogate 2):")
print(df_dmf['dmf_kmc_who_sur2'].value_counts(dropna=False))
print(df_dmf['dmf_kmc_who_sur2'].value_counts(normalize=True, dropna=False) * 100)
```

```python
# Other Intervention/Status Variables
print("\nDistribution of 'dmf_l12_mech_vent' (Mechanical Ventilation in last 12 hrs):")
print(df_dmf['dmf_l12_mech_vent'].value_counts(dropna=False))
print(df_dmf['dmf_l12_mech_vent'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_inotropes' (Inotropes in last 12 hrs):")
print(df_dmf['dmf_l12_inotropes'].value_counts(dropna=False))
print(df_dmf['dmf_l12_inotropes'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_ivflud' (IV Fluids in last 12 hrs):")
print(df_dmf['dmf_l12_ivflud'].value_counts(dropna=False))
print(df_dmf['dmf_l12_ivflud'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_l12_spsis_diag' (Sepsis Diagnosis in last 12 hrs):")
print(df_dmf['dmf_l12_spsis_diag'].value_counts(dropna=False))
print(df_dmf['dmf_l12_spsis_diag'].value_counts(normalize=True, dropna=False) * 100)
```

```python
# Feeding Patterns
print("\nDistribution of 'dmf_bf_bentfed' (Breastfeeding Initiated in last 12 hrs):")
print(df_dmf['dmf_bf_bentfed'].value_counts(dropna=False))
print(df_dmf['dmf_bf_bentfed'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_bf_formilk' (Formula Milk in last 12 hrs):")
print(df_dmf['dmf_bf_formilk'].value_counts(dropna=False))
print(df_dmf['dmf_bf_formilk'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_bf_bentfed_hw' (Breastfeeding by Healthcare Worker in last 12 hrs):")
print(df_dmf['dmf_bf_bentfed_hw'].value_counts(dropna=False))
print(df_dmf['dmf_bf_bentfed_hw'].value_counts(normalize=True, dropna=False) * 100)

print("\nDistribution of 'dmf_bf_bsucbrst' (Baby Sucking at Breast in last 12 hrs):")
print(df_dmf['dmf_bf_bsucbrst'].value_counts(dropna=False))
print(df_dmf['dmf_bf_bsucbrst'].value_counts(normalize=True, dropna=False) * 100)
```

```python
# Daily Record Count per Baby
daily_records_per_baby = df_dmf.groupby('dmf_babyid')['dmf_dof'].nunique()
print("\nStatistics on Number of Daily Records per Baby:")
print(daily_records_per_baby.describe())

# Display a few examples of babies with many records, and few records
print("\nExamples of babies with many daily records:")
print(daily_records_per_baby.nlargest(5))

print("\nExamples of babies with few daily records:")
print(daily_records_per_baby.nsmallest(5))
```

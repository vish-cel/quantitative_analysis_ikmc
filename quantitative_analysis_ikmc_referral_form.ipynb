```python
import pandas as pd
```

```python
# Load the CSV file, skipping the first 2 rows which are metadata
df_referral = pd.read_csv(r"C:\Users\CEL\Desktop\Quantitative Analysis\iKMC\Referral form_17th Sep '24 - 18th Jun '25.csv", skiprows=2)
```

```python
# Calculate descriptive statistics for numerical columns
print("\nDescriptive Statistics for Numerical Columns:")
print(df_referral.describe())
```

```python
# Check for missing values and their percentages
print("\nMissing Values and their Percentages:")
missing_values = df_referral.isnull().sum()
missing_percentage = (df_referral.isnull().sum() / len(df_referral)) * 100
missing_df = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage': missing_percentage})
print(missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing Percentage', ascending=False).to_markdown(numalign="left", stralign="left"))
```

```python
# Analyze unique values and their counts for a few selected categorical/identifier columns that seem relevant
print("\nUnique Values and Counts for selected columns:")
for col in ['country_id', 'site_id', 'rf_reason_1', 'rf_bsex_1', 'rf_trprt_1', 'rf_mc_sm', 'rf_mc_mrfrl', 'rf_mc_mrf_wb', 'rf_sngl_mltpl']:
    if col in df_referral.columns:
        print(f"\n--- {col} ---")
        print(df_referral[col].value_counts().to_markdown(numalign="left", stralign="left"))
```
